```python
import os
import cv2
from tqdm import tqdm

os.chdir('/content/drive/MyDrive/BOAZ_수어프로젝트/Data/10481_12994_2') # 총 동영상 338개
output_path = '/content/drive/MyDrive/BOAZ_수어프로젝트/Data/10481_12994_frame_2/' # 경로 마지막 '/' 필요


for video in tqdm(os.listdir('/content/drive/MyDrive/BOAZ_수어프로젝트/Data/10481_12994_2')):

    capture = cv2.VideoCapture(video)
    
    frameNr = 0

    # 동영상 이름 설정
    video_name_1 = list(map(lambda x: x.split('.')[0] ,video)) 
    video_name = ''.join(video_name_1[:-4])

    while (True):
    
        success, frame = capture.read()
    
        if success:
            cv2.imwrite(f'{output_path}{video_name}_{frameNr}.jpg', frame)

        else:
            break
    
        frameNr = frameNr+1
        
    
    capture.release()
```
